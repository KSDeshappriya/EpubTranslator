```python
from html.parser import HTMLParser
from googletrans import Translator, LANGUAGES
```

```python
# Sample HTML content
html_content = """
<!DOCTYPE html>
<html>
<body>
<h1> hi bro </h1>
<p>This is a paragraph.</p>
<p>This is another paragraph.</p>
</body>
</html>
"""
```

```python
class MyHTMLParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.translator = Translator()
        self.translated_text = []
        self.current_data = []

    def handle_data(self, data):
        # Handle text data within HTML elements
        self.current_data.append(data)

    def handle_endtag(self, tag):
        # When an HTML tag ends, join and translate the collected data
        if tag in {'h1', 'p'}:
            text = ' '.join(self.current_data)
            if text.strip():
                translated = self.translator.translate(text, src='auto', dest='si')
                self.translated_text.append(translated.text)
            self.current_data = []
```

```python
# Create an instance of the custom HTML parser
parser = MyHTMLParser()

# Feed the HTML content to the parser
parser.feed(html_content)

# save as translated txt in array
outedTxt = parser.translated_text
print(outedTxt)
```

```python
import re

def extract_text_between_tags(input_string):
    # Define a regular expression pattern to match HTML or XML tags
    tag_pattern = r'<[^>]+>'
    
    # Use the re.split() function to split the input string into text fragments using the tags as separators
    text_fragments = re.split(tag_pattern, input_string)
    
    # Filter out empty strings and whitespace
    text_fragments = [fragment.strip() for fragment in text_fragments if fragment.strip()]
    
    # Define a regular expression pattern to match URLs
    url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    
    # Initialize a list to store identified URLs
    detected_urls = []
    
    # Iterate through the text fragments and extract URLs
    for fragment in text_fragments:
        urls_in_fragment = re.findall(url_pattern, fragment)
        detected_urls.extend(urls_in_fragment)
    
    # Return the list of text fragments and detected URLs
    return text_fragments, detected_urls

# Example usage:
input_string = '''
<html>
  <head>
    <title>Sample Page</title>
  </head>
  <body>
    <p>This is a sample page with a URL: https://example.com and another URL: http://example.org.</p>
  </body>
</html>
'''

text_fragments, detected_urls = extract_text_between_tags(input_string)

if text_fragments:
    print("Text fragments between tags:")
    for fragment in text_fragments:
        print(fragment)
else:
    print("No text fragments found between tags.")

if detected_urls:
    print("\nDetected URLs:")
    for url in detected_urls:
        print(url)
else:
    print("No URLs detected.")
```

```python
import re
from ebooklib import epub
from deep_translator import GoogleTranslator
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm  # Progress bar
import time

def translate_epub(input_epub_path, output_epub_path, target_language='en', batch_size=10):
    # Open the input EPUB file
    book = epub.read_epub(input_epub_path)

    # Initialize the Google Translate API
    translator = GoogleTranslator(source='auto', target=target_language)

    def translate_text(content):
        text_fragments = extract_text_between_tags(content)
        translated_fragments = [translator.translate(fragment) for fragment in text_fragments]
        translated_content = content
        for i in range(len(text_fragments)):
            translated_content = translated_content.replace(text_fragments[i], translated_fragments[i], 1)
        return translated_content

    # Create a ThreadPoolExecutor for concurrent translation
    with ThreadPoolExecutor() as executor:
        start_time = time.time()
        for item in tqdm(book.items, desc="Translating EPUB"):
            if isinstance(item, epub.EpubHtml):
                content = item.content.decode('utf-8')
                translated_content = executor.submit(translate_text, content).result()
                item.content = translated_content.encode('utf-8')

        end_time = time.time()

    # Save the translated EPUB to the output file
    epub.write_epub(output_epub_path, book)

    # Calculate and display the execution time
    execution_time = end_time - start_time
    print(f"Translation completed in {execution_time:.2f} seconds.")

def extract_text_between_tags(input_string):
    tag_pattern = r'<[^>]+>'
    text_fragments = re.split(tag_pattern, input_string)
    text_fragments = [fragment.strip() for fragment in text_fragments if fragment.strip()]
    return text_fragments

if __name__ == "__main__":
    input_epub_path = 'input.epub'  # Replace with your input EPUB file path
    output_epub_path = 'output.epub'  # Replace with your desired output file path
    target_language = 'si'  # Change this to your desired target language code
    translate_epub(input_epub_path, output_epub_path, target_language)
```
